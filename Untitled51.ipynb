```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv("/content/train_u6lujuX_CVtuZ9i (1).csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.dropna()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
dataset['Loan_Status']=dataset['Loan_Status'].map({'Y':1,'N':0})
```

```python
dataset.head()
```

```python
dataset['Property_Area'].value_counts()
```

```python
dataset['Self_Employed']=dataset['Self_Employed'].map({'Yes':1,'No':0})
```

```python
dataset['Education']=dataset['Education'].map({'Graduate':1,'Not Graduate':0})
```

```python
dataset['Property_Area']=dataset['Property_Area'].map({'Urban':2,'Semiurban':1,'Rural':0})
```

```python
dataset['Gender']=dataset['Gender'].map({'Male':1,'Female':0})
```

```python
dataset['Married']=dataset['Married'].map({'Yes':1,'No':0})
```

```python
dataset.head()
```

```python
dataset['Dependents'].value_counts()
```

```python
dataset=dataset.replace(to_replace='3+',value=4)
```

```python
dataset['Dependents'].value_counts()
```

```python
dataset.head()
```

```python
dataset.drop(columns='Loan_ID',inplace=True)
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['Loan_Status']
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,stratify=y,random_state=42)
```

```python
from sklearn import svm
model=svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)
score=accuracy_score(y_pred,y_train)
```

```python
print(score)
```

```python
y_pred_test=model.predict(x_test)
score_test=accuracy_score(y_pred_test,y_test)
print(score_test)
```
